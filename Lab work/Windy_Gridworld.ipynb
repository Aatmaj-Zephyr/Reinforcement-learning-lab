```python
import numpy as np
grid_height = 10
grid_length = 7
wind_intensity = [0,0,0,1,1,1,2,2,1,0]
possible_actions = [(1,0),(0,1),(0,-1)] #right, up, down
start = (0,3)
goal = (7,3)
action_matrix= []


for state_x in range(grid_length):
    action_matrix.append([])
    for state_y in range(grid_height):
      action_matrix[state_x].append([])
      for action_num in range(len(possible_actions)):
        init_prob = 100

        action_matrix[state_x][state_y].append(init_prob)
action_matrix = np.array(action_matrix,dtype=np.float64)
print(action_matrix)
```

```python
import random
current_state = start

for k in range(10000):
  gamma = 0.99
  current_state =  (random.randint(0,grid_length-1),random.randint(0,grid_height-1))

  for i in range(1000):
    action = random.choices(possible_actions, weights= action_matrix[current_state[0]][current_state[1]])[0]
    action_index = possible_actions.index(action)
    x_move = max(min(current_state[0] + action[0],grid_length-1),0)
    y_move = max(min(current_state[1] + action[1]+wind_intensity[x_move],grid_height-1),0)

    current_state = (x_move,y_move)

    reward = -0.01*((x_move-goal[1]-1)**2+(y_move-goal[0])**2)

    action2 = random.choices(possible_actions, weights= action_matrix[current_state[0]][current_state[1]])[0]
    action_index2 = possible_actions.index(action2)
    x_move_2 = max(min(current_state[0] + action[0],grid_length-1),0)
    y_move_2 = max(min(current_state[1] + action[1],grid_height-1),0) # no wind here
    action_matrix[x_move][y_move][action_index] = action_matrix[x_move][y_move][action_index]+0.001*(reward+gamma*action_matrix[x_move_2][y_move_2][action_index2]-action_matrix[x_move][y_move][action_index])
    action_matrix[x_move][y_move][action_index] = max(action_matrix[x_move][y_move][action_index],0.00001)
     #print(action_matrix)

```

```python
best_actions = []
for x in range(grid_length):
  best_actions.append([])
  for y in range(grid_height):
    a = np.argmax(action_matrix[x][y])
    if( a==0):
      best_actions[x].append(">")
    elif(a==1):
      best_actions[x].append("^")
    else:
      best_actions[x].append("v")
best_actions = np.array(best_actions)
print(best_actions)
```
